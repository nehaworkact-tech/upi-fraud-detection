# Notebook 01: Business Understanding
### UPI Payment Fraud Detection System

## Problem Statement
UPI processes 14+ billion transactions monthly in India.
Fraud is the #1 threat to every bank, fintech and payment company.

This project builds an end-to-end fraud detection system that:
- Identifies fraudulent transactions instantly
- Explains why a transaction was flagged (RBI compliance)
- Quantifies business value saved
- Enables risk-based decisions: Block / Review / Allow

## Why Standard Accuracy Fails in Fraud Detection
In our dataset, only 0.17% of transactions are fraud.
A model that predicts "not fraud" every single time = 99.83% accurate.
But it catches ZERO frauds. Completely useless.

This is called the **Class Imbalance Problem.**
It changes everything about how we build and evaluate this model.

## The Cost Matrix
| | Actual: Fraud | Actual: Legit |
|---|---|---|
| **Predicted: Fraud** | ✅ True Positive — fraud caught | ❌ False Positive — customer blocked |
| **Predicted: Legit** | ❌ False Negative — fraud missed, money lost| ✅ True Negative — correct |

**False Negative** = fraud slips through = direct financial loss to bank

**False Positive** = genuine customer blocked = customer frustration, 
potential churn, reputational damage

These are NOT equal errors. A missed fraud costs more than a 
blocked legitimate transaction. Our model must reflect this.

## How We Will Measure Success
Not accuracy. Instead:
- **Precision:** Of transactions we flag as fraud, how many actually are?
- **Recall:** Of all actual frauds, how many did we catch?
- **F1 Score:** Harmonic mean of precision and recall
- **PR-AUC:** Precision-Recall curve — best metric for imbalanced data
- **Business Cost Saved:** Estimated ₹ value of fraud prevented

## Business Decision Framework
Every transaction gets one of three outcomes:
- 🔴 **Block** — high fraud probability (>80%) — stop immediately
- 🟡 **Review** — medium risk (40-80%) — send for manual review  
- 🟢 **Allow** — low risk (<40%) — approve instantly

*Note: Exact thresholds will be determined in Notebook 12 
based on the business cost matrix and Precision-Recall curves. 
Thresholds are never set before modelling in production.*

## Models We Will Build
### Supervised (learns from labelled fraud history)
- Logistic Regression — baseline
- LightGBM — fast, production-grade
- XGBoost + SHAP — best performance + explainability

### Unsupervised (detects anomalies without labels)
- Isolation Forest — catches unknown fraud patterns
- AutoEncoder — deep learning anomaly detection

### Deep Learning
- Neural Network — feedforward fraud classifier
